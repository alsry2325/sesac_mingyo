```python
#6-1.생존 여부와 pclass와의 관계  그룹화 hue매개변수
#분포 확인
sns.countplot(data=df,x="Pclass",hue="Survived")
```

```python
#Pclass와 생존 여부에 대한 비율 (Pclass와 사망에 대한비율)
dead_pclass= df[df["Survived"]==0]["Pclass"].value_counts()

plt.pie(x=dead_pclass.values.tolist(),labels=dead_pclass.index.tolist(),autopct="%.1f%%")
```

```python
survived_pclass= df[df["Survived"]==1]["Pclass"].value_counts()
#티켓클래스 당 생존자에대한 분포
plt.pie(x=survived_pclass.values.tolist(),labels=survived_pclass.index.tolist(),autopct="%.1f%%")
```

```python
#6-2 성별에 대한 생존자수
survived_sex= df[df["Survived"]==1]["Sex"].value_counts()

plt.countplot(x=survived_sex.values.tolist(),labels=survived_sex.index.tolist(),autopct="%.1f%%")
```

```python
#6-3. 나이에 따른 생존 여부 탐색
#나이의 생존 여부 분포
sns.histplot(data=df,x="Age",hue="Survived")
```

```python
tmp = df.dropna(subset=["Age","Survived"]).copy() # 나이 결측치 제거

#구간의 범위를 지정
bins = [0,10,20,30,40,50,60,70,100]
# 가 구간의 이름(라벨)
labels =["0~9","10~19","20~29","30~39","40~49","50~59","60~69","70+"]

tmp["AgeBin"] = pd.cut(tmp["Age"],bins=bins,labels=labels)
#cut() 함수 : 연속형 데이터를  구간(범주)으로 나누는 함수

#각 나이 구간마다의 평균 생존율%로 확인
#각 구간을 그룹화 ->평균->100 곱해서 백분율 형태
rate_by_bin = tmp.grouby("AgeBin")["Survived"].mean().mul(100).reset_index()

sns.barplot(data=rate_by_bin, x="AgeBin",y="Survived")

plt.xlabel("나이의 구간")
plt.ylabel("생존율(%)")
plt.ylim(0,100)
```
